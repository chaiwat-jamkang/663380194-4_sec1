# Lab 4.1: Student Management System (List Operations)

```python
# สร้าง List ว่าง
student_names = []

# เพิ่มรายชื่อนักเรียนเริ่มต้น 3-5 คน
num = int(input("ต้องการเพิ่มนักเรียนกี่คน (3-5 คน): "))

if 3 <= num <= 5:
    for i in range(num):
        name = input(f"กรอกชื่อนักเรียนคนที่ {i+1}: ")
        student_names.append(name)

    print("\nรายชื่อนักเรียนปัจจุบัน:", student_names)

    # เมนูหลัก
    while True:
        print("\n===== เมนู =====")
        print("1. เพิ่มรายชื่อนักเรียน")
        print("2. ลบรายชื่อนักเรียน")
        print("3. ค้นหารายชื่อนักเรียน")
        print("4. เรียงลำดับรายชื่อ")
        print("5. ออกจากโปรแกรม")

        choice = input("เลือกเมนู (1-5): ")

        # เพิ่ม
        if choice == "1":
            name = input("กรอกชื่อนักเรียนที่ต้องการเพิ่ม: ")
            student_names.append(name)
            print("เพิ่มเรียบร้อย")
            print("รายชื่อนักเรียน:", student_names)

        # ลบ
        elif choice == "2":
            name = input("กรอกชื่อนักเรียนที่ต้องการลบ: ")
            if name in student_names:
                student_names.remove(name)
                print("ลบเรียบร้อย")
            else:
                print("ไม่พบชื่อนักเรียน")
            print("รายชื่อนักเรียน:", student_names)

        # ค้นหา
        elif choice == "3":
            name = input("กรอกชื่อนักเรียนที่ต้องการค้นหา: ")
            if name in student_names:
                print(name, "อยู่ที่ตำแหน่ง Index =", student_names.index(name))
            else:
                print("ไม่พบชื่อนักเรียน")

        # เรียงลำดับ
        elif choice == "4":
            student_names.sort()
            print("เรียงลำดับแล้ว")
            print("รายชื่อนักเรียน:", student_names)

        # ออกจากโปรแกรม
        elif choice == "5":
            print("ออกจากโปรแกรม")
            break

        else:
            print("กรุณาเลือกเมนู 1-5 เท่านั้น")

else:
    print("กรุณากรอกจำนวนระหว่าง 3 ถึง 5 คน")
```
