```python
# รับข้อมูลจากผู้ใช้ และ แปลงข้อมูลเป็นจำนวนเต็ม (int)
user_input = input("กรุณากรอกตัวเลขจำนวนเต็ม: ")
number = int(user_input)

# และ (Challenge): ตรวจสอบว่าเป็นบวก/ลบ/ศูนย์ และเป็นคู่/คี่ พร้อมกัน
if number > 0:
    # ตรวจสอบคู่/คี่ สำหรับเลขบวก
    if number % 2 == 0:
        result = "เป็นจำนวนเต็มบวก และเป็นเลขคู่"
    else:
        result = "เป็นจำนวนเต็มบวก และเป็นเลขคี่"

elif number < 0:
    # ตรวจสอบคู่/คี่ สำหรับเลขลบ
    if number % 2 == 0:
        result = "เป็นจำนวนเต็มลบ และเป็นเลขคู่"
    else:
        result = "เป็นจำนวนเต็มลบ และเป็นเลขคี่"

else:
    # กรณีที่เป็นศูนย์
    result = "เป็นศูนย์ (และเป็นเลขคู่)"

# แสดงผลลัพธ์
print(f"ตัวเลข {number} คือ: {result}")
```
