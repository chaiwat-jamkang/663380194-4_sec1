# **Activity 3.1: Debugging Infinite Loops & Loop Logic Errors**

**Buggy Snippet A: Infinite Loop**

**โค้ดที่มีปัญหา**

```python
counter = 0

while counter < 5:
    print(counter)
    # Missing counter increment!
```

**ปัญหา**

- ตัวแปร counter ไม่เคยเพิ่มค่า

- เงื่อนไข counter < 5 จึงเป็นจริงตลอด

- ทำให้เกิด Infinite Loop

**วิธีแก้**

```python
counter = 0

while counter < 5:
    print(counter)
    counter += 1
```

**Buggy Snippet C: Misplaced Break**

**โค้ดที่มีปัญหา**



```python
for char in "hello":
    if char == 'l':
        print("Found 'l', stopping!")
        break
    print(char)
```

**ปัญหา**

ผลลัพธ์คือ

h

e

Found 'l', stopping!
\

ตัวอักษร l ไม่ถูกพิมพ์ เพราะ break ทำงานก่อน print(char)

**วิธีแก้ (ให้พิมพ์ตัวอักษรก่อน)**

```python
for char in "hello":
    print(char)
    if char == 'l':
        print("Found 'l', stopping!")
        break
```

**ผลลัพธ์**

h

e

l

Found 'l', stopping!

**4. Challenge: Pyramid**

โค้ด

```python
rows = 5

for i in range(rows):
    for j in range(rows - i - 1):
        print(" ", end="")
    for k in range(2 * i + 1):
        print("*", end="")
    print()
```
