# Lab 4.2: Geometric Calculations (Tuples & Immutability)

```python
import math

# กำหนดพิกัดจุด
point1 = (3, 4)
point2 = (6, 8)

# เข้าถึงค่าพิกัด
print("Point 1: x =", point1[0], ", y =", point1[1])
print("Point 2: x =", point2[0], ", y =", point2[1])

# ทดลองแก้ไขค่า (Tuple แก้ไขไม่ได้)
# point1[0] = 5
# จะเกิด TypeError เพราะ Tuple เป็นข้อมูลแบบ Immutable
# เมื่อสร้างแล้วไม่สามารถแก้ไขสมาชิกภายในได้

# คำนวณระยะห่าง
distance = math.sqrt((point2[0] - point1[0]) ** 2 + (point2[1] - point1[1]) ** 2)

print("ระยะห่างระหว่าง Point 1 และ Point 2 =", distance)
```

```python
import math

# รับค่าพิกัดจุดที่ 1
x1 = float(input("กรอกค่า x ของ Point 1: "))
y1 = float(input("กรอกค่า y ของ Point 1: "))

# รับค่าพิกัดจุดที่ 2
x2 = float(input("กรอกค่า x ของ Point 2: "))
y2 = float(input("กรอกค่า y ของ Point 2: "))

# เก็บเป็น Tuple
point1 = (x1, y1)
point2 = (x2, y2)

print("Point 1:", point1)
print("Point 2:", point2)

# คำนวณระยะห่าง
distance = math.sqrt((point2[0] - point1[0]) ** 2 + (point2[1] - point1[1]) ** 2)

print("ระยะห่างระหว่าง Point 1 และ Point 2 =", distance)
```
