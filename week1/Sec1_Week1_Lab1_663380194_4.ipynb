1.เขียนสคริปต์สั้น ๆ เพื่อแก้โจทย์ปัญหา (Word problems) เช่น การคำนวณพื้นที่สี่เหลี่ยม หรือการ
แปลงอุณหภูมิจากเซลเซียสเป็นฟาเรนไฮต์

```python
# กำหนดค่าความกว้างและความยาว
width = 5.0
height = 10.0

# คำนวณพื้นที่
area = width * height

# แสดงผลลัพธ์
print("พื้นที่สี่เหลี่ยมคือ:", area)
```

```python
# กำหนดอุณหภูมิเป็นเซลเซียส
celsius = 30.0

# คำนวณเป็นฟาเรนไฮต์
fahrenheit = (celsius * 9/5) + 32

# แสดงผลลัพธ์
print("อุณหภูมิในหน่วยฟาเรนไฮต์คือ:", fahrenheit)
```
