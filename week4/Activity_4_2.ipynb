# Activity 4.2: Interactive To-Do List (using List Operations)

```python
# สร้าง List ว่าง
todo_list = []

while True:
    print("\n===== To-Do List =====")
    print("1. เพิ่มรายการงาน")
    print("2. ดูรายการงานทั้งหมด")
    print("3. ทำเครื่องหมายว่างานเสร็จสิ้นแล้ว")
    print("4. ออกจากโปรแกรม")

    choice = input("เลือกเมนู (1-4): ")

    # เพิ่มรายการงาน
    if choice == "1":
        task = input("กรอกรายการงาน: ")
        todo_list.append(task)
        print("เพิ่มงานเรียบร้อย")

    # ดูรายการงานทั้งหมด
    elif choice == "2":
        if len(todo_list) == 0:
            print("ไม่มีรายการงาน")
        else:
            print("\nรายการงานทั้งหมด")
            for i in range(len(todo_list)):
                print(i, "-", todo_list[i])

    # ลบงานที่ทำเสร็จแล้ว
    elif choice == "3":
        if len(todo_list) == 0:
            print("ไม่มีรายการงาน")
        else:
            print("\nรายการงาน")
            for i in range(len(todo_list)):
                print(i, "-", todo_list[i])

            index = int(input("กรอกหมายเลขงานที่ทำเสร็จ: "))

            if 0 <= index < len(todo_list):
                removed = todo_list.pop(index)
                print("ทำงานเสร็จแล้ว:", removed)
            else:
                print("หมายเลขไม่ถูกต้อง")

    # ออกจากโปรแกรม
    elif choice == "4":
        print("ออกจากโปรแกรม")
        break

    else:
        print("กรุณาเลือกเมนู 1-4")
```
