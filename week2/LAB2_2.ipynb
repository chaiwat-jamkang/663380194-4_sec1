```python
# รับอายุและแปลงเป็นจำนวนเต็ม
age_input = input("กรุณากรอกอายุของคุณ: ")
age = int(age_input)

# (Challenge) ถามความชอบเพิ่มเติมเพื่อนำไปคิดเงื่อนไขร่วม
action_lover = input("คุณชอบภาพยนตร์แนวแอ็กชัน (Action) หรือไม่? (ใช่/ไม่ใช่): ")

# แนะนำเรทภาพยนตร์ โดยใช้ if-elif-else ร่วมกับเงื่อนไข and
if age >= 18 and action_lover == "ใช่":
    result = "แนะนำเรท R หรือ 18+ (คุณสามารถดูหนังแอ็กชันที่มีความรุนแรงหรือฉากต่อสู้สมจริงได้)"

elif age >= 18 and action_lover != "ใช่":
    result = "แนะนำเรท 18+ แนวทั่วไป (เช่น ดราม่า หรือไซไฟ ที่เนื้อหาเข้มข้นแต่ไม่ใช่สายลุย)"

elif age >= 13 and action_lover == "ใช่":
    result = "แนะนำเรท PG-13 (คุณสามารถดูหนังแอ็กชันฮีโร่ หรือหนังผจญภัยทั่วไปได้สนุก)"

elif age >= 13 and action_lover != "ใช่":
    result = "แนะนำเรท PG-13 แนวครอบครัว/คอมเมดี้ (เนื้อหาเหมาะสมกับวัยรุ่น)"

else:
    # กรณีอายุต่ำกว่า 13 ปี ไม่ว่าจะชอบแอ็กชันหรือไม่ ก็ควรดูเรททั่วไป
    result = "แนะนำเรท G หรือ PG (ภาพยนตร์ทั่วไป เหมาะสำหรับทุกวัยหรือควรมีผู้ปกครองให้คำแนะนำ)"

# แสดงผลลัพธ์
print(f"\nผลการวิเคราะห์สำหรับอายุ {age} ปี:")
print(f"-> {result}")
```
