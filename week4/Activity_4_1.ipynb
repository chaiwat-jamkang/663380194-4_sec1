# Activity 4.1: List vs. Tuple: Scenario Matching

1. ข้อมูลอุณหภูมิประจำวันที่มีการบันทึกไว้ตลอดหนึ่งสัปดาห์ ควรใช้ List

เหตุผล เพราะอาจมีการแก้ไขหรือเพิ่มข้อมูลอุณหภูมิย้อนหลังได้

2. ค่าสี RGB ของเฉดสีเฉพาะเฉดหนึ่ง ควรใช้ Tuple

เหตุผล เพราะค่า RGB ของสีหนึ่ง ๆ เป็นค่าคงที่ ไม่ควรเปลี่ยนแปลง

3. รายการสินค้าในรถเข็นชอปปิงที่สามารถกดเพิ่มหรือลบออกได้ ควรใช้ List

เหตุผล เพราะมีการเพิ่ม ลบ หรือแก้ไขรายการสินค้าได้ตลอดเวลา

4. รายชื่อวันในหนึ่งสัปดาห์เรียงตามลำดับ ควรใช้ Tuple

เหตุผล เพราะลำดับและจำนวนวันในหนึ่งสัปดาห์เป็นข้อมูลคงที่ ไม่เปลี่ยนแปลง

5. คะแนนเก็บของนักเรียนจากการส่งงานชุดต่าง ๆ ซึ่งอาจมีการอัปเดตหรือแก้ไขคะแนนในภายหลังได้ ควรใช้ List

เหตุผล เพราะคะแนนสามารถแก้ไขหรืออัปเดตได้

6. พิกัด GPS ของสถานที่สำคัญหรือจุดแลนด์มาร์กที่ตั้งอยู่กับที่ ควรใช้ Tuple

เหตุผล เพราะพิกัดของสถานที่เป็นข้อมูลคงที่ ไม่ควรแก้ไข
